```python
import matplotlib.pyplot as plt

categorical_columns = ['Dx:Cancer', 'Dx:CIN', 'Dx:HPV', 'Dx', 'Hinselmann', 'Schiller', 'Citology', 'Biopsy']
fig, axes = plt.subplots(nrows=len(categorical_columns), ncols=1, figsize=(10, 20))

for i, col in enumerate(categorical_columns):
    cancer_df[col].value_counts().plot(kind='bar', ax=axes[i], title=col)
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel('Category')

plt.tight_layout()
plt.show()
```

```python
numerical_columns = [
    'Age', 'Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies', 
    'Smokes', 'Smokes (years)', 'Smokes (packs/year)', 'Hormonal Contraceptives', 
    'Hormonal Contraceptives (years)', 'IUD', 'IUD (years)', 'STDs', 'STDs (number)',
    'STDs:condylomatosis', 'STDs:cervical condylomatosis', 'STDs:vaginal condylomatosis', 
    'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis', 'STDs:pelvic inflammatory disease', 
    'STDs:genital herpes', 'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV', 
    'STDs:Hepatitis B', 'STDs:HPV', 'STDs: Number of diagnosis', 'STDs: Time since first diagnosis', 
    'STDs: Time since last diagnosis'
]

fig, axes = plt.subplots(nrows=len(numerical_columns) // 3 + 1, ncols=3, figsize=(15, 30))
axes = axes.flatten() 

for i, col in enumerate(numerical_columns):
    cancer_df[col].hist(bins=15, ax=axes[i])
    axes[i].set_title(col)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Explore the relationship between numerical variables and categorical data where the categorical values are '1' (positive cases)

```python
positive_cancer_df = cancer_df[cancer_df['Dx:Cancer'] == 1]  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example to visualize the age distribution for positive 'Dx:Cancer'
sns.boxplot(x='Dx:Cancer', y='Age', data=cancer_df)
plt.title('Age Distribution by Cancer Diagnosis')
plt.show()
```

```python
grouped = cancer_df.groupby('Dx:Cancer')['Age'].describe()
print(grouped)
```

```python
from scipy.stats import ttest_ind, mannwhitneyu

# Testing age difference between Dx:Cancer positive and negative
group1 = cancer_df[cancer_df['Dx:Cancer'] == 1]['Age']
group2 = cancer_df[cancer_df['Dx:Cancer'] == 0]['Age']

# T-test
t_stat, p_val = ttest_ind(group1, group2)
print("T-test p-value:", p_val)

# Mann-Whitney U test for non-parametric data
u_stat, p_val_u = mannwhitneyu(group1, group2)
print("Mann-Whitney U test p-value:", p_val_u)
```
